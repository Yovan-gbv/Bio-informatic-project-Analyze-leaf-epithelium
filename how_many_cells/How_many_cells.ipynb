# How many cells 

##### In this question we want to determine the number of cells in each image. So we have to make the segmentation of each picture then label the mask and finally count the number of labels.


### All the imports 

```python
import numpy as np
import pandas as pd
import glob
from src.fonctions_I import*
```

### Parameters

```python
path_longues = 'longues/*.tif'
path_puzzle = 'puzzle/*.tif'
path_croco = 'croco/*.tif'
path_stomates = 'stomates/*.tif'
path_veines = 'veines/*.tif'
```

## Cases to run

## Dataframe Longues cells

```python
image_fname_list = []
nb_cell_list =[]
centroids_fname_list = []
dico_data = {'image_fname' : image_fname_list, 'nb_cells' : nb_cell_list, 'centroids_fname' : centroids_fname_list}


for i in glob.glob(path_longues) :
    #Do the segmentation for all the images for this category 
    label = segmentation_longue(i)
    #Splitted the "/" so we can have at each / the word is in the next 
    #line and take the last string (which correspond to the image name)
    splitted_path = i.split("/")
    image_fname_list.append(splitted_path[-1])
    #Do a dataframe with all the centroids for each image and convert it to csv file 
    props = measure.regionprops_table(
    label,
    properties=['label', 'centroid'] 
    )

    df_props = pd.DataFrame(props)

    centroids = df_props[['centroid-0','centroid-1']] 
    centroids_csv_data = df_props.to_csv('cells_centroid_' + splitted_path[-1].replace("tif","csv"), index = True)
    centroids_fname_list.append('cells_centroid_' + splitted_path[-1].replace("tif","csv"))

    #Count the number of cells 
    unique_values = np.unique(label)
    nb_colony = len(unique_values)

    dico_data = {'image_fname' : image_fname_list, 'nb_cells' : nb_cell_list, 'centroids_fname' : centroids_fname_list}

    nb_cell_list.append(nb_colony)

    #Convert the dic to a dataframe 
    df_longues = pd.DataFrame(dico_data)

df_longues
```
